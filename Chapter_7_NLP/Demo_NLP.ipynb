### **Encoder**

```python
text = 'The cat sat on the mat. The dog ate my hoomework'

words = text_to_word_sequence(text)
vocabs = set(words)
print('words: \n', words)
print('len of words', len(words))
print('vocabs: \n', vocabs)
vocab_size = len(vocabs)

print('len vocab: ', vocab_size)

#integer encode the document
# list số nguyên từ 1-> n, mỗi số nguyên mã hóa 1 từ, ko đảm bảo tính duy nhất
res = one_hot(text, round(vocab_size*1.3)) # 1.3 là cái số ngta tìm ra là ổn nhất á
print(res)
```

### **Tokenier**

```python
from tensorflow.keras.preprocessing.text import Tokenizer

docs = ['Well done!',
        'Good job',
        'Great effort',
        'Nice job abc abc',
        'Excellent abc!']
```

```python
t = Tokenizer()
t.fit_on_texts(docs)
```

```python
print(t.word_counts)
print(t.document_count)
print(t.word_index)
print(t.word_docs)
```

```python
encoded_docs = t.texts_to_matrix(docs, mode='count')
encoded_docs
```

```python
'''
vì index của text dc đánh từ 1->n
mà array có index từ 0->n-1

=> tạo ra cột đầu toàn bằng 0, chỉ mảng tính chất thỏa index text từ 1->n hoyy
'''
```

### **Transformation**

```python
reviews = ['movie was great', 'really bad', 'i loved it', 'actor is handsome', 's2',
           'the best movie ever', 'could be better', 'mer']
```

```python
t1 = Tokenizer()
t1.fit_on_texts(reviews)
```

```python
x = t1.texts_to_sequences(reviews)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
x = pad_sequences(x, maxlen=3, truncating='post')
```
