## **Linear Model**
- Cơ bản mà nói thì đối vs hàm Y = a*x + b thì có thể áp dụng phương pháp ở dưới:

```python
## **Linear Model**
# cthuc tổng quan của simple linear regression
def linear_regression(intercept, slope, feature):
    return intercept + feature * slope

# mse
def loss_function(intercept, slope, targets,feature):
    predictions = linear_regression(intercept, slope, feature)
    return tf.keras.losses.mse(targets,predictions)
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('iris')
```

```python
sns.regplot(data.petal_length, data.petal_width)
plt.show()
```

```python
np.array(data.petal_length, np.float32)
```

```python
tf.cast(data.petal_length, tf.float32)
```

```python
length = np.array(data.petal_length, np.float32) # giống tf.cast(data.petal_length, tf.float32)
width = np.array(data.petal_width, np.float32)

intercept = tf.Variable(0.1, np.float32) # giá trị sẽ thay đổi => dùng Variable
slope = tf.Variable(0.1, np.float32)
```

```python
# tối ưu hóa
opt = tf.keras.optimizers.Adam()
```

```python
opt
```

```python
for j in range(1000):
    opt.minimize(
        lambda: loss_function(intercept, slope, length, width),
        var_list=[intercept, slope],
    )
    # print(loss_function(intercept, slope, length, width)) # mse
```

```python
# MSE:
tf.print(loss_function(intercept, slope, length, width))
```
