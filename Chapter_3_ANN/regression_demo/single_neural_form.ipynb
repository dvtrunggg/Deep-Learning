```python
def get_value(tf_x):
    with tf.Session() as sess:
        return sess.run(tf_x)
```

```python
model = keras.Sequential([layers.Dense(units=1, input_shape=[1])])

x = tf.linspace(-1, 1, 100)
y = model.predict(x)

plt.plot(x, y, "k")
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel('Input x')
plt.ylabel('Target y')
w, b = model.weights # <=> get_weights()

title = 'weight: '+ str(w[0][0])+ 'bias: '+ str(b[0])
# plt.title(title)
plt.show()
```
